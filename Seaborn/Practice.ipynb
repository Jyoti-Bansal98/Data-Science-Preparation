# Seaborn Practice

```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

# correlation with heatmap

```python
df.select_dtypes(include = 'number').corr()
```

```python
 sns.heatmap(df.select_dtypes(include = 'number').corr())
```

# JoinPlot

```python
sns.jointplot(x = 'tip',y = 'total_bill', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'tip',y = 'total_bill',data=df, kind = 'reg')
```

# Pair plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

# Dist plot

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde = False, bins = 10)
```

# count plot

```python
sns.countplot(x = 'sex', data = df)
```

# Bar plot

```python
sns.barplot(x = 'total_bill', y = 'sex', data = df)
```

# Box plot

```python
sns.boxplot(x= 'smoker',y= 'total_bill', data = df)
```

```python
sns.boxplot(x= 'day',y= 'total_bill', data=df, palette = 'rainbow')
```

```python
sns.boxplot(data = df, orient = 'v', palette = 'rainbow')
```

```python
sns.boxplot(x = 'total_bill', y = 'day',hue = 'smoker', data = df)
```

# Violin plot

```python
sns.violinplot(x = 'total_bill', y = 'day', data = df,
               palette = 'rainbow')
```
