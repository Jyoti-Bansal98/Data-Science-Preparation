```python
df = pd.read_csv("climate_data.csv", parse_dates = ["day"])
type(df.day[0])
df.set_index('day', inplace = True)
df
```

#### fill NaN

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temp': 0,
    'windspeed': 0,
    'event': 'no event'
  })
new_df
```

```python
new_df = df.fillna(method = 'bfill', axis = 'rows', limit = 1)
new_df
```

#### Interpolate

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna(how = 'all')
new_df
```

```python
new_df = df.dropna(thresh = 2)    #atleast 2 value ho baki na chalegi
new_df
```

#### Replace Function

```python
new_df = df.replace(-9999, np.nan)
new_df
```

```python
df = pd.read_csv("weather2_data.csv")
df
```

```python
new_df = df.replace([-9999,-8888, 'no event'],np.nan)
new_df
```

```python
new_df = df.replace({
    'temp': [-9999,-8888],
    'wind speed': -9999,
    'event': 'no event'
 },np.nan)
new_df
```

```python
new_df = df.replace({
      -9999: np.nan,
      -8888: np.nan,
      'no event': 'sunny'
    })
new_df
```

```python
new_df = df.replace({
    'temp': '[A-Za-z]',
    'wind speed': '[A-Za-z]'
   },'', regex = True )
new_df
```
